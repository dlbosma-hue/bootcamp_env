**Step 1: Setting Up and Loading Data**

```python

import numpy as np                                   
import pandas as pd                                   
import matplotlib.pyplot as plt                       
from sklearn.datasets import load_breast_cancer      
from sklearn.model_selection import train_test_split 
from sklearn.neighbors import KNeighborsClassifier   
from sklearn.metrics import (accuracy_score,         
                            precision_score, 
                            recall_score, 
                            confusion_matrix, 
                            classification_report)

print("✓ All libraries imported successfully!\n")

# STEP 2: LOAD THE BREAST CANCER DATASET
print("Loading breast cancer dataset...")
cancer_data = load_breast_cancer()


print(f"\nDataset type: {type(cancer_data)}")
print(f"Number of samples: {len(cancer_data.data)}")  # How many patients' data?
print(f"Number of features: {len(cancer_data.feature_names)}")  # How many measurements?
print(f"Target classes: {cancer_data.target_names}")  # What are we predicting? 

# STEP 3: CONVERT TO DATAFRAME 

df = pd.DataFrame(cancer_data.data, columns=cancer_data.feature_names)
df['target'] = cancer_data.target  # Add the target column

print("\n" + "="*60)
print("FIRST FEW ROWS OF DATA")
print("="*60)
print(df.head())

print("\n" + "="*60)
print("DATASET INFO")
print("="*60)
print(df.info())

print("\n" + "="*60)
print("TARGET DISTRIBUTION (How many of each type?)")
print("="*60)
print(df['target'].value_counts())
print(f"Benign (0) - Healthy: {(df['target'] == 0).sum()} samples")
print(f"Malignant (1) - Cancer: {(df['target'] == 1).sum()} samples")
```

**Step 2: Data Exploration**

```python
# STEP 2: EXPLORE THE DATA
print("\n" + "="*60)
print("BASIC STATISTICS (Understanding the data)")
print("="*60)
print(df.describe())
# This shows: count, mean (average), std (spread), min, 25%, 50%, 75%, max

# STEP 3: CHECK FOR MISSING VALUES
print("\n" + "="*60)
print("MISSING VALUES CHECK")
print("="*60)
missing = df.isnull().sum()
if missing.sum() == 0:
    print("✓ No missing values found!")
    print("  (All 569 patients have complete data)")
else:
    print("Missing values found:")
    print(missing[missing > 0])

# STEP 4: VISUALIZE KEY FEATURES
print("\n" + "="*60)
print("CREATING VISUALIZATIONS")
print("="*60)

# Looking at 4 key measurements
key_features = ['mean radius', 'mean texture', 'mean perimeter', 'mean area']

# Create 4 subplots (2 rows, 2 columns)
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
axes = axes.ravel()  # Flatten to 1D array for easier looping

for idx, feature in enumerate(key_features):
    # Histogram for benign (0) - blue
    axes[idx].hist(df[df['target'] == 0][feature], 
                   alpha=0.5, label='Benign', bins=30, color='blue')
    # Histogram for malignant (1) - orange
    axes[idx].hist(df[df['target'] == 1][feature], 
                   alpha=0.5, label='Malignant', bins=30, color='orange')
    
    axes[idx].set_xlabel(feature, fontsize=10)
    axes[idx].set_ylabel('Frequency (how many)', fontsize=10)
    axes[idx].set_title(f'Distribution of {feature}', fontsize=11, fontweight='bold')
    axes[idx].legend()

plt.tight_layout()
plt.savefig('feature_distributions.png', dpi=150, bbox_inches='tight')
print("✓ Saved visualization to 'feature_distributions.png'")
plt.show()
```
