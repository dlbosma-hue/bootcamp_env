```python
def should_trigger_security_alert(sensor, reading_value, system_mode):

    sensor_type = sensor["type"]
    
    # Security alerts only for motion and door sensors
    if sensor_type == "motion" or sensor_type == "door":
        # Only alert if the system is in AWAY mode
        if system_mode == "AWAY":
            # Check if the sensor detected something abnormal
            if is_abnormal_reading(sensor, reading_value):
                return True  # Yes, trigger security alert!
            else:
                return False  # False alarm, reading is normal
        else:
            return False  # Not in AWAY mode, no security alert
    else:
        return False  # Not a motion/door sensor, no security alert
    
```

*Functions*

```python
def generate_reading(sensor):
    """
    Generates a realistic reading for a sensor.
    
    Parameters:
    - sensor: Sensor dictionary
    
    Returns:
    - A realistic reading value based on sensor type
    """
    sensor_type = sensor["type"]
    
    if sensor_type == "temperature":
        # Generate a random temperature between 30-100°F
        return random.randint(30, 100)
    
    elif sensor_type == "motion":
        # Randomly return True (motion detected) or False (no motion)
        # 20% chance of motion, 80% chance of no motion
        return random.random() < 0.2
    
    elif sensor_type == "door":
        # Randomly return "OPEN" or "CLOSED"
        # 10% chance of open, 90% chance of closed
        if random.random() < 0.1:
            return "OPEN"
        else:
            return "CLOSED"
    
    elif sensor_type == "smoke":
        # Randomly return "CLEAR" or "DETECTED"
        # 5% chance of smoke detected, 95% chance of clear
        if random.random() < 0.05:
            return "DETECTED"
        else:
            return "CLEAR"
    
    return None  # Unknown sensor type
```

```python
def process_reading(sensor, reading_value, system_mode):
    """
    Processes a sensor reading and determines if alerts are needed.
    
    Parameters:
    - sensor: Sensor dictionary
    - reading_value: The reading from the sensor
    - system_mode: Current system mode (HOME, AWAY, SLEEP)
    
    Returns:
    - A list of alerts (empty if no alerts needed)
    """
    alerts = []  # Start with an empty list
    timestamp = datetime.now().strftime("%H:%M:%S")  # Get current time
    
    # ========== SECURITY ALERTS ==========
    # Check for motion/door triggers while in AWAY mode
    if should_trigger_security_alert(sensor, reading_value, system_mode):
        if sensor["type"] == "motion":
            alert = create_alert(
                "HIGH",
                f"SECURITY: Motion detected in {sensor['location']} while in {system_mode} mode!",
                sensor["id"],
                timestamp
            )
            alerts.append(alert)
        
        elif sensor["type"] == "door":
            alert = create_alert(
                "HIGH",
                f"SECURITY: {sensor['location']} opened while in {system_mode} mode!",
                sensor["id"],
                timestamp
            )
            alerts.append(alert)
    
    # ========== SAFETY ALERTS ==========
    # Check for dangerous temperatures
    if sensor["type"] == "temperature":
        if reading_value < 35:
            alert = create_alert(
                "CRITICAL",
                f"SAFETY: Temperature in {sensor['location']} is {reading_value}°F - FROZEN PIPE RISK!",
                sensor["id"],
                timestamp
            )
            alerts.append(alert)
        
        elif reading_value > 95:
            alert = create_alert(
                "CRITICAL",
                f"SAFETY: Temperature in {sensor['location']} is {reading_value}°F - EQUIPMENT FAILURE RISK!",
                sensor["id"],
                timestamp
            )
            alerts.append(alert)
    
    # Check for smoke
    if sensor["type"] == "smoke":
        if reading_value == "DETECTED":
            alert = create_alert(
                "CRITICAL",
                f"SAFETY: SMOKE DETECTED in {sensor['location']} - FIRE RISK!",
                sensor["id"],
                timestamp
            )
            alerts.append(alert)
    
    # ========== COMFORT NOTIFICATIONS ==========
    # Only in HOME mode, check if temperature is uncomfortable
    if system_mode == "HOME" and sensor["type"] == "temperature":
        if reading_value < 65 or reading_value > 75:
            alert = create_alert(
                "LOW",
                f"COMFORT: Temperature in {sensor['location']} is {reading_value}°F (comfort range is 65-75°F)",
                sensor["id"],
                timestamp
            )
            alerts.append(alert)
    
    return alerts  # Return the list of alerts (could be empty)
 
```

```python
def trigger_alert(alert):
    """
    Displays an alert to the user with nice formatting and emojis.
    
    Parameters:
    - alert: Alert dictionary
    """
    severity_symbol = {
        "LOW": "ℹ️",
        "MEDIUM": "⚠️",
        "HIGH": "🚨",
        "CRITICAL": "🔥"
    }
    
    symbol = severity_symbol.get(alert["severity"], "⚠️")
    print(f"[ALERT!] {symbol} {alert['severity']}: {alert['message']}")
```

```python
def log_event(message, timestamp=None):
    """
    Logs an event to the console.
    
    Parameters:
    - message: The message to log
    - timestamp: Optional timestamp (uses current time if not provided)
    """
    if timestamp is None:
        timestamp = datetime.now().strftime("%H:%M:%S")
    print(f"[LOG] [{timestamp}] {message}")
```

*Classes* 

```python
class Sensor:
    """
    Represents a sensor in the HomeGuard system.
    """
    
    def __init__(self, sensor_id, location, sensor_type, threshold=None):
        """
        Initializes a new sensor (the constructor).
        
        This function runs automatically when you create a new Sensor object.
        
        Parameters:
        - sensor_id: Unique identifier for the sensor
        - location: Where the sensor is located
        - sensor_type: Type of sensor ("motion", "temperature", "door", "smoke")
        - threshold: Optional threshold value for the sensor
        """
        # Store all the information as properties of this object
        self.id = sensor_id              # Use 'self' to make this belong to this specific sensor
        self.location = location
        self.type = sensor_type
        self.threshold = threshold
        self.current_value = None        # No reading yet
    
    def read(self):
        """
        Generates and stores a new reading for this sensor.
        
        Returns:
        - The reading value
        """
        # Generate a realistic reading
        if self.type == "temperature":
            self.current_value = random.randint(30, 100)
        elif self.type == "motion":
            self.current_value = random.random() < 0.2
        elif self.type == "door":
            self.current_value = "OPEN" if random.random() < 0.1 else "CLOSED"
        elif self.type == "smoke":
            self.current_value = "DETECTED" if random.random() < 0.05 else "CLEAR"
        
        return self.current_value
    
    def isAbnormal(self):
        """
        Checks if the current reading is abnormal.
        
        Returns:
        - True if abnormal, False if normal
        """
        if self.type == "temperature":
            if self.current_value is None:
                return False
            return self.current_value < 35 or self.current_value > 95
        
        elif self.type == "motion":
            return self.current_value == True
        
        elif self.type == "door":
            return self.current_value == "OPEN"
        
        elif self.type == "smoke":
            return self.current_value == "DETECTED"
        
        return False
    
    def reset(self):
        """
        Resets the sensor's current reading to None.
        """
        self.current_value = None
    
    def __str__(self):
        """
        Returns a string representation of the sensor.
        This is what prints when you do print(sensor_object)
        """
        status = "No reading" if self.current_value is None else str(self.current_value)
        return f"{self.id} ({self.location}): {status}"
        
```

*Integration*

```python
def run_simulation(duration_minutes=5, system_mode="AWAY"):
    """
    Runs the HomeGuard security system simulation.
    
    Parameters:
    - duration_minutes: How long to run the simulation (default: 5 minutes)
    - system_mode: System mode (HOME, AWAY, SLEEP)
    """
    print("=" * 50)
    print("=== HomeGuard Security System ===")
    print("=" * 50)
    print(f"Mode: {system_mode}\n")
    
    # Create sensor objects
    sensors = [
        Sensor("MOTION_001", "Living Room", "motion"),
        Sensor("TEMP_001", "Kitchen", "temperature", threshold=35),
        Sensor("DOOR_001", "Front Door", "door"),
        Sensor("SMOKE_001", "Bedroom", "smoke")
    ]
    
    # Simulate time passing (each iteration = 1 minute)
    for minute in range(duration_minutes):
        current_time = datetime.now().strftime("%H:%M:%S")
        print(f"\nTime: {current_time}")
        
        # Read all sensors
        for sensor in sensors:
            reading = sensor.read()  # Generate a new reading
            
            # Display the reading in a nice format
            if sensor.type == "temperature":
                status = "Normal" if 65 <= reading <= 75 else "Abnormal"
                print(f"[READING] {sensor.location} Temperature: {reading}°F ({status})")
            
            elif sensor.type == "motion":
                status = "DETECTED" if reading else "No activity"
                print(f"[READING] {sensor.location} Motion: {status}")
            
            elif sensor.type == "door":
                print(f"[READING] {sensor.location}: {reading}")
            
            elif sensor.type == "smoke":
                print(f"[READING] {sensor.location} Smoke: {reading}")
            
            # Process the reading and trigger alerts
            # Convert sensor object to dict format for process_reading function
            sensor_dict = {
                "id": sensor.id,
                "location": sensor.location,
                "type": sensor.type,
                "threshold": sensor.threshold
            }
            alerts = process_reading(sensor_dict, reading, system_mode)
            
            # Display all alerts
            for alert in alerts:
                trigger_alert(alert)
                # If it's serious, log that we're sending a notification
                if alert["severity"] in ["HIGH", "CRITICAL"]:
                    log_event("Sending notification to homeowner...")
        
        # Small delay to make output readable (optional)
        import time
        time.sleep(0.5)  # Wait 0.5 seconds before next iteration
    
    # Simulation complete
    print("\n" + "=" * 50)
    print("Simulation complete!")
    print("=" * 50)
```

```python
# Main execution
if __name__ == "__main__":
    # Run the simulation with AWAY mode for 3 minutes
    run_simulation(duration_minutes=3, system_mode="AWAY")
```
