Create 3 helper functions:
  - call_openai() - Sends prompt, gets response
  - test_prompt_multiple_times() - Runs same prompt N times
  - analyze_results() - Measures consistency

```python
def call_openai(prompt, system_message="You are a helpful assistant.", temperature=0.7, max_tokens=None, seed=None):
    """Helper function to call OpenAI API with specified parameters"""
    try:
        params = {
            "model": MODEL,
            "messages": [
                {"role": "system", "content": system_message},
                {"role": "user", "content": prompt}
            ],
            "temperature": temperature
        }
        
        if max_tokens:
            params["max_tokens"] = max_tokens
        if seed is not None:
            params["seed"] = seed
            
        response = client.chat.completions.create(**params)
        return response.choices[0].message.content
    except Exception as e:
        return f"Error: {str(e)}"

def display_response(title, response, params=None):
    """Display a response with formatting"""
    print(f"\n{'='*60}")
    print(f"ðŸ“ {title}")
    if params:
        print(f"Parameters: {params}")
    print(f"{'='*60}")
    print(response)
    print(f"{'='*60}\n")

def count_tokens_approx(text):
    """Approximate token count (roughly 4 characters per token)"""
    return len(text) // 4

print("âœ… Helper functions loaded!")
```

```python
def test_prompt_multiple_times(prompt, num_runs=5, delay=1):
    """
    Test the same prompt multiple times to measure consistency
    
    Parameters:
    - prompt: the prompt to test
    - num_runs: number of times to run 
    - delay: seconds to wait between runs
    
    Returns: list of all responses
    """
    results = []
    
    for i in range(num_runs):
        print(f"  Running iteration {i+1}/{num_runs}...", end=" ")
        response = call_openai(prompt)
        results.append(response)
        print("Done")
        time.sleep(delay)
    
    return results

print("\nHelper Function test_prompt_multiple_times()")
print("âœ“ Function created successfully")
```

```python
def analyze_results(results, task_name="Task"):
    """
    Analyze consistency of results from multiple runs
    
    Parameters:
    - results: list of responses from test_prompt_multiple_times()
    - task_name: name of the task
    
    Returns: dictionary with analysis metrics
    """
    print(f"\n{'='*70}")
    print(f"ANALYSIS: {task_name}")
    print(f"{'='*70}")
    print(f"Total runs: {len(results)}\n")
    print("Responses (first 150 chars):")
    print("-" * 70)
    
    for i, result in enumerate(results, 1):
        preview = result[:150] + "..." if len(result) > 150 else result
        print(f"Run {i}: {preview}")
    
    # Calculate metrics
    unique_responses = set(results)
    consistency = (len(results) - len(unique_responses)) / len(results) * 100
    
    print(f"\n{'-'*70}")
    print(f"Total unique responses: {len(unique_responses)} out of {len(results)}")
    print(f"Consistency: {consistency:.1f}%")
    print(f"{'='*70}\n")
    
    return {
        "total_runs": len(results),
        "unique_responses": len(unique_responses),
        "consistency": consistency
    }

print("\nHelper Function analyze_results()")
print("âœ“ Function created successfully")
```


**REFLECTION**

**What failure patterns?**


v1 was vague, so the model interpreted it liberally. For Sentiment Analysis, without format rules, it gave varied outputâ€”sometimes just the word "POSITIVE", sometimes explanations mixed in. That's 0% consistency. For Product Description, I literally just said "write a description for this product" with no constraints. The model produced outputs ranging from 1-5 paragraphs with different tones (sometimes poetic, sometimes technical). No rules meant no consistency. Data Extraction was interesting because v1 actually had a baseline of 53.3% consistency, which means it got the right format sometimes by luck. But when it didn't, it cycled through three different formats: prose paragraphs, bullet points, and key-value pairs. The pattern was clear: without explicit format specifications or examples, the model just picked reasonable-looking options. It wasn't brokenâ€”the prompts were just too open-ended.


**Which techniques had the biggest impact?**


Few-shot examples delivered the largest improvements. Sentiment jumped from 80% (v2 with just constraints) to 95% (v3 with examples). Same pattern for Product: 70% to 85%. For Extraction: 85% to 95%. The key insight is that showing the model three correct examples works better than describing the format. Structured output (the 4-part format: Opening, Features, Benefit, Call to Action) was the second-biggest win because it created a container.(This was a process, where I learned also along with some help) Instead of the model deciding how to organize information, it just filled in pre-defined sections. Chain-of-Thought for Data Extraction forced systematic thinking; when I made the model show its reasoning steps (Name â†’ Email â†’ Product Model â†’ Issue), consistency jumped from 85% to 95%. But technique impact varies by task complexity. CoT added huge value for a multi-step task like Extraction, but for simple classification like Sentiment, the improvement was minimal. 

**What I would do next time**


First, I wasted steps. I didn't test sarcasm or mixed sentiment until like Step 12. I should have tried variations way earlier to find the edge cases when I was still building v1 or v2. Second, I didn't have a target metric at the start. I just kept improving until things looked good. Next time I'd say "I want 95% consistency" right at the beginning and measure v1 as baseline so I could actually see my progress clearly. Third, I'd think about the task type before writing or trying out ANY prompts to be fair. Sentiment is simple (one word), Product is creative (needs structure), Extraction is complex (needs reasoning). Knowing that upfront would tell me which technique to focus on instead of trying everything. Finally, I would actually document why my examples worked. I used three examples and they worked great, but I didn't think about: Did they show clear patterns? Did they cover different scenarios? That would help me make better examples for other tasks instead of just guessing.
